```python
def find_outliers(food_production):
    # Calculate quartiles and interquartile range (IQR)
    Q1 = food_production.quantile(0.25)
    Q3 = food_production.quantile(0.75)
    IQR = Q3 - Q1

    # Calculate lower and upper fences for outlier detection
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR

    # Identify outliers using boolean indexing
    outliers = food_production[(food_production < lower_fence) | (food_production > upper_fence)]

    return outliers
```

```python
outliers = find_outliers(food_production['Value'])

print ("Max Outlier Value : "+ str(outliers.max()))
print ("Number Of Outliers : "+ str(len(outliers)))
print(outliers)
```

```python
def find_outliers(food_supply):
    # Calculate quartiles and interquartile range (IQR)
    Q1 = food_supply.quantile(0.25)
    Q3 = food_supply.quantile(0.75)
    IQR = Q3 - Q1

    # Calculate lower and upper fences for outlier detection
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR

    # Identify outliers using boolean indexing
    outliers = food_supply[(food_supply < lower_fence) | (food_supply > upper_fence)]

    return outliers
```

```python
outliers = find_outliers(food_supply['Value'])

print ("Max Outlier Value : "+ str(outliers.max()))
print ("Number Of Outliers : "+ str(len(outliers)))
print(outliers)
```

```python
food_production_two=pd.DataFrame(food_production.groupby('Year')['Value'].mean()).reset_index(inplace= False)
food_production_two
```

```python
food_production_two.rename(columns={"Value": "average_food_produced"}, inplace=True)
food_production_two
```

```python
food_supply_two=pd.DataFrame(food_supply.groupby('Year')['Value'].mean()).reset_index(inplace= False)
food_supply_two
```

```python
food_supply_two.rename(columns={"Value": "average_food_supplied"}, inplace=True)
food_supply_two
```

```python
foodproduction_and_supply = pd.merge(food_production_two, food_supply_two, on='Year')
foodproduction_and_supply
```

```python
fig= plt.figure(figsize=(20,10))

ax = fig.add_subplot() 
ax.plot(foodproduction_and_supply['Year'],foodproduction_and_supply['average_food_produced'],marker='d', label="Food produced", color='#00ff00')
ax.plot(foodproduction_and_supply['Year'],foodproduction_and_supply['average_food_supplied'],marker='o', label="Food supplied", color='#800080')

ax.set_ylabel('Values')
ax.set_title('Average Food Production vs Supply for Each Year')
ax.set_xlabel('Year')
ax.legend()
plt.grid()
fig.savefig('production vs supply.jpg')
plt.show()
```
